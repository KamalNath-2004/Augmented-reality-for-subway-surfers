```python
def detect_pose(image, pose, draw, display):
    output_img = image.copy()
    imageRGB = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    result = pose.process(imageRGB)

    if result.pose_landmarks and draw:
        mp_drawing.draw_landmarks(
            image=output_img,
            landmark_list=result.pose_landmarks,
            connections=mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=mp_drawing.DrawingSpec(color=(255, 255, 255), thickness=3, circle_radius=3),
            connection_drawing_spec=mp_drawing.DrawingSpec(color=(49, 125, 237), thickness=3, circle_radius=2)
        )

    if display:
        plt.figure(figsize=[22, 22])
        plt.subplot(121)
        plt.imshow(image[:, :, ::-1])
        plt.title("Real Image")
        plt.axis('off')
        plt.subplot(122)
        plt.imshow(output_img[:, :, ::-1])
        plt.title("Output Image")
        plt.axis('off')

    return output_img, result
```

```python
from math import hypot
def checkHandsJoined(image , result , draw = False , display = False):
    height , width = image.shape[:2]
    output_image = image.copy()
    
    left_wrist_landmark = (
    result.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST].x * width,
    result.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST].y * height
)

    right_wrist_landmark = (
        result.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_WRIST].x * width,
        result.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_WRIST].y * height
    )

    eulidean_distance=int(hypot(left_wrist_landmark[0]-right_wrist_landmark[0],
                                left_wrist_landmark[1]-right_wrist_landmark[1]))
    if eulidean_distance < 270:
        hand_status =  "Hands Joined"
        color = (0, 225, 0)
    else:
        hand_status = "Hands Not Joined" 
        color=(0 ,0 ,225)  
        
    if draw:
        cv2.putText(output_image,hand_status,(10, 30),cv2.FONT_HERSHEY_PLAIN, 2, color, 3 )
        cv2.putText(output_image,f'Distance:{eulidean_distance}',(10 ,70,),
                    cv2.FONT_HERSHEY_PLAIN, 2 , 3)
    if display:
        plt.figure(figsize=[10,10])
        plt.imshow(output_image[:,:,::-1]);plt.title("Output Image");plt.axis('off')
    else:
        return output_image,hand_status    
                  
```

```python
def checkLeftRight(image, result, draw=False, display=False):
    horizontal_pos = None
    height, width = image.shape[:2]
    output_img = image.copy()
    
    left_x = int(result.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].x * width)
    right_x = int(result.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].x * width)
    
    if right_x <= width // 2 and left_x <= width // 2:
        horizontal_pos = 'Left'
    elif right_x >= width // 2 and left_x >= width // 2:
        horizontal_pos = 'Right'
    elif right_x >= width // 2 and left_x <= width // 2:
        horizontal_pos = 'Center'
    
    if draw:
        cv2.putText(output_img, f'Position: {horizontal_pos}', (10, 30), cv2.FONT_HERSHEY_PLAIN, 2, (0, 255, 0), 2)

        if display:
            plt.figure(figsize=[10, 10])
            plt.imshow(output_img[:, :, ::-1])
            plt.title("Output Image")
            plt.axis('off')

    return output_img, horizontal_pos  
```

```python
def checkJumpCrouch(image, results, MID_Y=250, draw=False, display=False):
    height, width = image.shape[:2]
    output_img = image.copy()
    
    left_y = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].y * height)
    right_y = int(results.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].y * height)
    
    actual_mid_y = abs(left_y + right_y) // 2

    # Adjusted bounds to make crouch less sensitive
    lower_b = MID_Y - 40  # More room to detect jumping
    upper_b = MID_Y + 20  # Crouch only when significantly below MID_Y

    if actual_mid_y < lower_b:
        posture = 'Jumping'
    elif actual_mid_y > upper_b:
        posture = 'Crouching'
    else:
        posture = 'Idle'

    if draw:
        cv2.putText(output_img, posture, (5, height - 50), cv2.FONT_HERSHEY_PLAIN, 2, (255, 255, 255), 3)
        cv2.line(output_img, (0, MID_Y), (width, MID_Y), (255, 255, 255), 2)

    if display:
        plt.figure(figsize=[10, 10])
        plt.imshow(output_img[:, :, ::-1])
        plt.title("Output Image")
        plt.axis("off")
    else:
        return output_img, posture
```

```python
import cv2
import time as t
import numpy as np
#from your_pose_utils_file import detect_pose, checkHandsJoined, classify_pose, checkJumpCrouch  #  Make sure it's imported

# Thresholds
SIDE_THRESHOLD = 50  # Only side threshold remains here

def main():
    cap = cv2.VideoCapture(0)
    p_time = t.time()

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        frame = cv2.flip(frame, 1)
        frame, result = detect_pose(frame, draw=True, display=False)

        gesture = "None"
        jump_status = "Normal"
        side_status = "Center"

        if result.pose_landmarks:
            hands = checkHandsJoined(frame, result)
            gesture = hands if hands else "Unknown"

            # Get landmarks
            landmarks = result.pose_landmarks.landmark
            left_hip = landmarks[23]
            right_hip = landmarks[24]
            nose = landmarks[0]

            img_h, img_w, _ = frame.shape
            center_y = int((left_hip.y + right_hip.y) * img_h / 2)
            center_x = int((left_hip.x + right_hip.x) * img_w / 2)

            #  Use your jump function here
            jump_status = checkJumpCrouch(nose, center_y, img_h)

            # Side detection
            if nose.x * img_w < center_x - SIDE_THRESHOLD:
                side_status = "Left"
            elif nose.x * img_w > center_x + SIDE_THRESHOLD:
                side_status = "Right"

            # Draw thresholds
            cv2.line(frame, (0, center_y), (img_w, center_y), (0, 255, 255), 1)
            cv2.line(frame, (center_x - SIDE_THRESHOLD, 0), (center_x - SIDE_THRESHOLD, img_h), (0, 0, 255), 1)
            cv2.line(frame, (center_x + SIDE_THRESHOLD, 0), (center_x + SIDE_THRESHOLD, img_h), (0, 0, 255), 1)
            cv2.line(frame, (center_x, 0), (center_x, img_h), (255, 255, 255), 1)
            cv2.line(frame, (0, int(nose.y * img_h)), (img_w, int(nose.y * img_h)), (255, 255, 255), 1)

        # Labels
        cv2.putText(frame, f"Jump: {jump_status}", (10, 40), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 255), 2)
        cv2.putText(frame, f"Side: {side_status}", (10, 80), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 255), 2)
        cv2.putText(frame, f"Gesture: {gesture}", (10, 120), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 127), 2)

        # FPS calculation
        c_time = t.time()
        fps = 1 / (c_time - p_time) if (c_time - p_time) != 0 else 0
        p_time = c_time
        cv2.putText(frame, f"FPS: {int(fps)}", (10, 160), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

        # Legend box
        legend_y = img_h - 100
        cv2.rectangle(frame, (10, legend_y), (300, legend_y + 90), (50, 50, 50), -1)
        cv2.putText(frame, "Legend:", (20, legend_y + 25), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255, 255, 255), 1)
        cv2.putText(frame, "Jumping / Crouching", (20, legend_y + 45), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 255), 1)
        cv2.putText(frame, "Left / Right / Center", (20, legend_y + 65), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 255), 1)
        cv2.putText(frame, "Gesture Detected", (20, legend_y + 85), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 127), 1)

        cv2.imshow("Pose Detection Interface", frame)

        if cv2.waitKey(5) & 0xFF == 27:
            break

    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```

```python
import cv2
import pyautogui
import time
import mediapipe as mp
from math import hypot

#  MediaPipe Setup 
mp_pose     = mp.solutions.pose
pose_img    = mp_pose.Pose(static_image_mode=True, 
                           min_detection_confidence=0.5, 
                           model_complexity=1)
pose_video  = mp_pose.Pose(static_image_mode=False, 
                           model_complexity=1, 
                           min_detection_confidence=0.7, 
                           min_tracking_confidence=0.7)
mp_drawing  = mp.solutions.drawing_utils

#  POSE & GESTURE FUNCTIONS 

def detect_pose(image, pose, draw=True, display=False):
    """Runs MediaPipe pose on `image`. Returns (output_image, result)."""
    output_img = image.copy()
    img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    result  = pose.process(img_rgb)

    if result.pose_landmarks and draw:
        mp_drawing.draw_landmarks(
            output_img,
            result.pose_landmarks,
            mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=mp_drawing.DrawingSpec((255,255,255), 3, 3),
            connection_drawing_spec=mp_drawing.DrawingSpec((49,125,237), 3, 2)
        )
    return output_img, result

def checkHandsJoined(image, result, draw=False):
    """Returns (image, 'Hands Joined'|'Hands Not Joined')."""
    h, w = image.shape[:2]
    output = image.copy()
    lw = result.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST]
    rw = result.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_WRIST]
    dist = int(hypot((lw.x-rw.x)*w, (lw.y-rw.y)*h))
    status = "Hands Joined" if dist < 200 else "Hands Not Joined"
    if draw:
        color = (0,225,0) if dist < 270 else (0,0,225)
        cv2.putText(output, status, (10,30), cv2.FONT_HERSHEY_PLAIN, 2, color, 3)
        cv2.putText(output, f"Dist:{dist}", (10,70), cv2.FONT_HERSHEY_PLAIN, 2, 3)
    return output, status

def checkLeftRight(image, result, draw=False):
    """Returns (image, 'Left'|'Right'|'Center')."""
    h, w = image.shape[:2]
    output = image.copy()
    ls = result.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER]
    rs = result.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER]
    lx, rx = int(ls.x*w), int(rs.x*w)
    if lx < w//2 and rx < w//2:
        status = "Left"
    elif lx > w//2 and rx > w//2:
        status = "Right"
    else:
        status = "Center"
    if draw:
        cv2.putText(output, f"Side: {status}", (10,30), cv2.FONT_HERSHEY_PLAIN, 2, (0,255,0), 2)
    return output, status

def checkJumpCrouch(image, result, MID_Y, draw=False):
    """Returns (image, 'Jumping'|'Crouching'|'Idle')."""
    h, w = image.shape[:2]
    output = image.copy()
    ls = int(result.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].y * h)
    rs = int(result.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].y * h)
    avg_y = (ls + rs) // 2

    lower_b = MID_Y - 40
    upper_b = MID_Y + 60

    if avg_y < lower_b:
        status = "Jumping"
    elif avg_y > upper_b:
        status = "Crouching"
    else:
        status = "Idle"

    if draw:
        cv2.putText(output, status, (5, h-50), cv2.FONT_HERSHEY_PLAIN, 2, (255,255,255), 3)
        cv2.line(output, (0, MID_Y), (w, MID_Y), (255,255,255), 2)
    return output, status


def main():
    cap = cv2.VideoCapture(0)
    cap.set(3, 1280)
    cap.set(4, 960)
    if not cap.isOpened():
        print("Camera failed to open.")
        return

    cv2.namedWindow("Gesture Interface", cv2.WINDOW_NORMAL)
    p_time = time.time()

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame")
            break

        frame = cv2.flip(frame, 1)
        frame, result = detect_pose(frame, pose_video, draw=True)

        h, w = frame.shape[:2]
        MID_Y    = h // 2
        SIDE_X   = w // 2

        # Default
        hand_status  = "None"
        side_status  = "Center"
        jump_status  = "Idle"

        if result.pose_landmarks:
            _, hand_status = checkHandsJoined(frame, result)
            _, side_status = checkLeftRight(frame, result)
            _, jump_status = checkJumpCrouch(frame, result, MID_Y)

        # Draw threshold lines
        cv2.line(frame, (0, MID_Y), (w, MID_Y), (255,255,255), 2)
        cv2.line(frame, (SIDE_X, 0), (SIDE_X, h), (255,255,255), 2)

        # Display statuses
        cv2.putText(frame, f"Jump: {jump_status}", (10,40), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,255), 2)
        cv2.putText(frame, f"Side: {side_status}", (10,80), cv2.FONT_HERSHEY_SIMPLEX, 1, (255,0,255), 2)
        cv2.putText(frame, f"Hands: {hand_status}", (10,120), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,127), 2)

        # FPS counter
        c_time = time.time()
        fps    = 1 / (c_time - p_time) if c_time != p_time else 0
        p_time = c_time
        cv2.putText(frame, f"FPS: {int(fps)}", (10,160), cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 2)

        # Legend box at bottom
        legend_y = h - 100
        cv2.rectangle(frame, (10, legend_y), (350, legend_y + 100), (50,50,50), -1)
        cv2.putText(frame, "Legend:",                   (20, legend_y + 25), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (255,255,255), 1)
        cv2.putText(frame, "Jumping / Crouching",       (20, legend_y + 45), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,255), 1)
        cv2.putText(frame, "Left / Right / Center",     (20, legend_y + 65), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,0,255), 1)
        cv2.putText(frame, "Hands Joined Status",       (20, legend_y + 85), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,127), 1)

        cv2.imshow("Gesture Interface", frame)

        # Exit on ESC
        if cv2.waitKey(5) & 0xFF == 27:
            break

    cap.release()
    cv2.destroyAllWindows()

if __name__ == "__main__":
    main()
```

```python
import cv2
import pyautogui
import time
import mediapipe as mp
from math import hypot
import subprocess
import threading
import tkinter as tk
from tkinter import filedialog, messagebox

mp_pose    = mp.solutions.pose
pose_video = mp_pose.Pose(static_image_mode=False,
                         model_complexity=1,
                         min_detection_confidence=0.7,
                         min_tracking_confidence=0.7)
mp_drawing = mp.solutions.drawing_utils

def detect_pose(image, pose, draw=True):
    out = image.copy()
    rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    res = pose.process(rgb)
    if res.pose_landmarks and draw:
        mp_drawing.draw_landmarks(
            out, res.pose_landmarks, mp_pose.POSE_CONNECTIONS,
            landmark_drawing_spec=mp_drawing.DrawingSpec((255,255,255),3,3),
            connection_drawing_spec=mp_drawing.DrawingSpec((49,125,237),3,2)
        )
    return out, res

def checkHandsJoined(res):
    lw = res.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_WRIST]
    rw = res.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_WRIST]
    dist = hypot((lw.x-rw.x), (lw.y-rw.y))
    return "Hands Joined" if dist < 0.1 else "Hands Not Joined"

def checkLeftRight(res):
    ls = res.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER]
    rs = res.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER]
    if ls.x < 0.5 and rs.x < 0.5:
        return "Left"
    if ls.x > 0.5 and rs.x > 0.5:
        return "Right"
    return "Center"

def checkJumpCrouch(res, mid_y=0.5):
    ls = res.pose_landmarks.landmark[mp_pose.PoseLandmark.LEFT_SHOULDER].y
    rs = res.pose_landmarks.landmark[mp_pose.PoseLandmark.RIGHT_SHOULDER].y
    avg = (ls+rs)/2
    if avg < mid_y - 0.1:
        return "Jumping"
    if avg > mid_y + 0.1:
        return "Crouching"
    return "Idle"



_running = False

def _controller_loop():
    global _running
    cap = cv2.VideoCapture(0)
    prev = None
    while _running:
        ret, frame = cap.read()
        if not ret: break
        frame = cv2.flip(frame,1)
        h, w = frame.shape[:2]
        mid_y = 0.5 

        frame, res = detect_pose(frame, pose_video)
        if res.pose_landmarks:
            hands = checkHandsJoined(res)
            side  = checkLeftRight(res)
            jump  = checkJumpCrouch(res, mid_y)

            # map to keys
            if hands=="Hands Joined" and prev!="space":
                pyautogui.press("space"); prev="space"
            elif jump=="Jumping" and prev!="up":
                pyautogui.press("up"); prev="up"
            elif jump=="Crouching" and prev!="down":
                pyautogui.press("down"); prev="down"
            elif side=="Left" and prev!="left":
                pyautogui.press("left"); prev="left"
            elif side=="Right" and prev!="right":
                pyautogui.press("right"); prev="right"
            elif jump=="Idle":
                prev=None

        cv2.imshow("Gesture Control", frame)
        if cv2.waitKey(1)==27:
            _running=False
            break

    cap.release()
    cv2.destroyAllWindows()

def start_controller():
    global _running
    if _running: return
    _running = True
    threading.Thread(target=_controller_loop, daemon=True).start()

def stop_controller():
    global _running
    _running = False


def launch_notepad():
    try:
        subprocess.Popen(["notepad.exe"])
    except FileNotFoundError:
        messagebox.showerror("Error", "Could not launch Notepad.")

def launch_game():
    path = filedialog.askopenfilename(
        title="Select Game EXE",
        filetypes=[("Executable", "*.exe")])
    if path:
        subprocess.Popen([path])

root = tk.Tk()
root.title("Gesture  Keyboard Controller")
root.geometry("300x200")

btn_notepad = tk.Button(root, text="Open Notepad", width=20, command=launch_notepad)
btn_notepad.pack(pady=5)

btn_game = tk.Button(root, text="Open Game...", width=20, command=launch_game)
btn_game.pack(pady=5)

btn_start = tk.Button(root, text="Start Controller", width=20, command=start_controller)
btn_start.pack(pady=5)

btn_stop = tk.Button(root, text="Stop Controller", width=20, command=stop_controller)
btn_stop.pack(pady=5)

root.protocol("WM_DELETE_WINDOW", lambda: (stop_controller(), root.destroy()))
root.mainloop()
```
